```python
from gurobipy import GRB
import gurobipy as gb
import pandas as pd
import numpy as np
```

```python
# Create the optimization model
question_1_model = gb.Model("Question 1: EcoGreen Energy")
```

```python
supply_df = pd.read_csv(r"C:\Users\gabri\Downloads\ecogreen_energy_supply.csv")
demand_df = pd.read_csv(r"C:\Users\gabri\Downloads\ecogreen_energy_demand.csv")
```

```python
demand_df
```

```python
demand = demand_df['Demand'].values
```

```python
variable_costs = supply_df.iloc[:, 3:]
variable_costs
```

```python
fixed_costs = supply_df['Fixed'].values
capacity = supply_df['Capacity'].values
```

```python
x = question_1_model.addVars(20, lb=0, vtype=GRB.BINARY, name="Sites")
y = question_1_model.addVars(20, 10, lb=0, vtype=GRB.CONTINUOUS, name="Electricity")
```

```python
fixed = gb.quicksum(fixed_costs[i]*x[i] for i in range(20))
variable = gb.quicksum(variable_costs.iloc[i,n]*y[i,n] for i in range(20) for n in range(10))
question_1_model.setObjective(fixed + variable, GRB.MINIMIZE)
```

```python
for i in range(20):
    question_1_model.addConstr(gb.quicksum(y[i,n] for n in range(10)) <= capacity[i]*x[i], "Capacity Constraint")
```

```python
for n in range(10):
    question_1_model.addConstr(gb.quicksum(y[i,n] for i in range(20)) >= demand[n], "Demand Constraint")
```

```python
for i in [9]:
    for j in [14,19]:
        question_1_model.addConstr(x[i] <= 1 - x[j], "Constraint 1")
```

```python
question_1_model.addConstr(2*x[3] <= x[3] + x[4], "Constraint 2")
```

```python
question_1_model.addConstr(x[4] <= x[7] + x[8], "Constraint 3")
```

```python
question_1_model.addConstr(gb.quicksum(x[i] for i in range(10)) <= 2 * (gb.quicksum(x[i] for i in range(10,20))), "Constraint 4")
```

```python
question_1_model.addConstr(gb.quicksum(y[i,n] for i in range(5) for n in range(10)) >= 0.3 * (gb.quicksum(y[j,n] for j in range(20) for n in range(10))), "Constraint 5")
```

```python
for n in range(10):
    for i in range(20):
        question_1_model.addConstr(y[i,n] <= 0.5 * (gb.quicksum(y[j,n] for j in range(20))), "Constraint 6")
```

```python
# Optimally solve the problem
question_1_model.optimize()
```

```python
# Value of the objective function
print("Cost: ", round(question_1_model.objVal,2))
```

```python
# Print the decision variables
print(question_1_model.printAttr('X'))
```

```python
# Get the optimal values of decision variables
electricity = question_1_model.getAttr('x', y)
electricity

for key, value in electricity.items():
    if value != 0:
        print(f"Site {key[0] + 1}, Province {key[1] + 1}: {value}")
```

```python
province_energy_count = {}

for key, value in electricity.items():
    if value != 0:
        province_index = key[1] + 1
        if province_index not in province_energy_count:
            province_energy_count[province_index] = 1
        else:
            province_energy_count[province_index] += 1

for province, count in province_energy_count.items():
    print(f"Province {province} receives energy from {count} sites.")
```

```python
num_decision_variables = question_1_model.numVars
print("Number of decision variables:", num_decision_variables)
```

```python
# Get the optimal objective function value
optimal_value = question_1_model.objVal

# Define a threshold for 1% difference
threshold = 0.01 * optimal_value

# Initialize variables to store the count and objective function values of feasible solutions within the threshold
count_within_threshold = 0
feasible_solution_values = []

# Iterate through all feasible solutions
for solution in question_1_model.getConstrs():
    # Check if the solution is feasible
    if solution.getAttr("slack") == 0.0:
        # Get the objective function value of the feasible solution
        solution_value = question_1_model.ObjVal
        # Check if the objective function value is within 1% of the optimal value
        if abs(solution_value - optimal_value) <= threshold:
            count_within_threshold += 1
            feasible_solution_values.append(solution_value)

# Print the number of feasible solutions within 1% of the optimal solution and their objective function values
print(f"Number of feasible solutions within 1% of the optimal solution: {count_within_threshold}")
print("Objective function values of the feasible solutions:")
for value in feasible_solution_values:
    print(value)
```
