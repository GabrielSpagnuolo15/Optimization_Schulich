```python
from gurobipy import GRB
import gurobipy as gb
import pandas as pd
import numpy as np
```

```python
# Create the optimization model
model = gb.Model("Question 1")
```

```python
nutrition_requirements_df = pd.read_csv(r"C:\Users\gabri\Downloads\nutrient_requirements.csv")
food_categories_df = pd.read_csv(r"C:\Users\gabri\Downloads\food_categories.csv")
food_preferences_df = pd.read_csv(r"C:\Users\gabri\Downloads\food_preferences.csv")
nutrition_content_df = pd.read_csv(r"C:\Users\gabri\Downloads\nutrient_content.csv")
```

```python
food_cost_per_gram = food_categories_df['Cost_per_gram'].values
food_cost_per_gram
```

```python
# x is the amount of food in grams
x = model.addVars(120, lb=0, vtype=GRB.CONTINUOUS, name="Food in Grams")
```

```python
# The objective function
objective = gb.quicksum(food_cost_per_gram[i]*x[i] for i in range(120))
model.setObjective(objective, GRB.MINIMIZE)
```

```python
nutrition_content_df = nutrition_content_df.drop(nutrition_content_df.columns[0], axis=1)
nutrition_content_df
```

```python
min_nutrition = nutrition_requirements_df['Min_Requirement'].values
min_nutrition
```

```python
max_nutrition = nutrition_requirements_df['Max_Requirement'].values
max_nutrition
```

```python
nutrient_totals = nutrition_content_df.sum(axis=0)
nutrient_totals
```

```python
for j in range(60):  # Loop over nutrient categories
    model.addConstr(gb.quicksum(nutrition_content_df.iloc[i,j] * x[i] for i in range(120)) >= min_nutrition[j])
    model.addConstr(gb.quicksum(nutrition_content_df.iloc[i,j] * x[i] for i in range(120)) <= max_nutrition[j])
```

```python
veg = food_categories_df['Is_Vegetarian'].values
veg
```

```python
veg_grams = food_preferences_df['Veggie_grams'].values
veg_grams
```

```python
veg_variety  = model.addConstr(gb.quicksum(veg[i]*x[i] for i in range (120)) >= veg_grams, "Veggie Preferences")
```

```python
vegan = food_categories_df['Is_Vegan'].values
vegan
vegan_grams = food_preferences_df['Vegan_grams'].values
vegan_grams
vegan_variety = model.addConstr(gb.quicksum(vegan[i]*x[i] for i in range (120)) >= vegan_grams, "Vegan Preferences")
```

```python
kosher = food_categories_df['Is_Kosher'].values
kosher
kosher_grams = food_preferences_df['Kosher_grams'].values
kosher_grams
kosher_variety = model.addConstr(gb.quicksum(kosher[i]*x[i] for i in range (120)) >= kosher_grams, "Kosher Preferences")
```

```python
halal = food_categories_df['Is_Halal'].values
halal
halal_grams = food_preferences_df['Halal_grams'].values
halal_grams
halal_variety = model.addConstr(gb.quicksum(halal[i]*x[i] for i in range (120)) >= halal_grams, "Halal Preferences")
```

```python
all = food_preferences_df['All_grams'].values
all
```

```python
model.addConstr(gb.quicksum(veg[i]*x[i] for i in range (120)) + gb.quicksum(vegan[i]*x[i] for i in range (120)) + gb.quicksum(kosher[i]*x[i] for i in range (120)) + gb.quicksum(halal[i]*x[i] for i in range (120)) >= all, "Vegan Preferences")
```

```python
for i in range(120):
    variety = model.addConstr(x[i] <= 0.03 * gb.quicksum(x[j] for j in range(120)))
```

```python
model.optimize()
```

```python
# Print the decision variables
print(model.printAttr('X'))
```

```python
# Value of the objective function
print("Cost of the Diet: ", round(model.objVal, 2))
```

```python
num_constraints = model.NumConstrs
print("Number of constraints (excluding non-negativity):", num_constraints)
```

```python
grams = model.getAttr('X', x)
grams
```

```python
print("")
print(f"Sensitivity Information for Land Capacity Constraint {variety.pi:.2f}:")
print("(LHS, RHS, Slack): ", (model.getRow(variety).getValue(), variety.RHS, variety.slack))
print("Shadow Price: ", variety.pi)
print("Range of Feasibility: ", (variety.SARHSUp, variety.SARHSLow))
```
